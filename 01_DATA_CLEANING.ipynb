```python
import pandas as pd
from datetime import datetime
def process_date(x):
    # Case 1: already datetime → swap
    if isinstance(x,(pd.Timestamp, datetime)):
        try:
            return x.replace(day=x.month, month=x.day)
        except:
            return pd.NaT
    # Case 2: string → just convert
    else:
        return pd.to_datetime(x, errors='coerce')
#read all files separately    
confirmed=pd.read_excel(r"C:\Users\sonali\Downloads\covid_19_dataset.xlsx",sheet_name="covid_19_confirmed_v1")
recovered=pd.read_excel(r"C:\Users\sonali\Downloads\covid_19_dataset.xlsx",sheet_name="covid_19_recovered_v1")
deaths_long=pd.read_excel(r"C:\Users\sonali\Downloads\deathsnew.xlsx")
#fill all provinces
confirmed['Province/State']=confirmed['Province/State'].fillna("All Provinces")
recovered['Province/State']=recovered['Province/State'].fillna("All Provinces")
confirmed=confirmed.drop(columns=['Lat','Long'],errors='ignore')
recovered=recovered.drop(columns=['Lat','Long'],errors='ignore')
#melt confirmed and recovered datasets
confirmed_long=confirmed.melt(id_vars=['Province/State','Country/Region'],var_name='Date',value_name='Confirmed')
recovered_long=recovered.melt(id_vars=['Province/State','Country/Region'],var_name='Date',value_name='Recovered')
#Convert to date column to appropriate date time
confirmed_long['Date']=confirmed_long['Date'].apply(process_date)
recovered_long['Date']=recovered_long['Date'].apply(process_date)
confirmed_long.to_excel(r"C:\Users\sonali\Downloads\confirmednew.xlsx",index=False)
recovered_long.to_excel(r"C:\Users\sonali\Downloads\recoverednew.xlsx",index=False)
```
